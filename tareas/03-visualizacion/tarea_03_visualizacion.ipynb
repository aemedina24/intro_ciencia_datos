```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dataset de trabajo
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
```

## Parte 1 — Figura y ejes

**P1.** Crea una figura con `plt.subplots()` de tamaño 8×5.  
Haz un gráfico de línea con los primeros 20 valores de `sepal_length`.  
Agrega título `"Longitud del sépalo"` y etiquetas en ambos ejes.

```python

fig, ax= plt.subplots(figsize=(8, 5))
ax.plot(df['sepal_length'].iloc[0:20], marker='o', linestyle='-', color="#E0A51B")


ax.set_xticks(range(0, 21, 2))

#Agregar título y etiquetas
ax.set_title("Longitud del sépalo", fontsize=14)
ax.set_xlabel("Indice")
ax.set_ylabel("Longitud (cm)")

# Mostrar el gráfico
```

## Parte 2 — Gráfico de barras

**P2.** Calcula el promedio de `petal_length` por especie (`species`) y guárdalo en una variable.  
Luego crea un **gráfico de barras vertical** con esos promedios.  
Agrega título y etiquetas en ambos ejes.

```python

mean_petal_lenght = df.groupby('species')['petal_length'].mean()


fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(mean_petal_lenght.index, mean_petal_lenght.values, color=['#3498db', '#2ecc71', '#e74c3c'])


ax.set_title("Promedio de Longitud del Pétalo por Especie", fontsize=14)
ax.set_xlabel("Especie")
ax.set_ylabel("Longitud Promedio (cm) petalo")


```

## Parte 3 — Dispersión (scatter plot)

**P3.** Crea un **gráfico de dispersión** (scatter) con `sepal_length` en el eje X y `sepal_width` en el eje Y.  
Usa un color diferente para cada especie (puedes filtrar el DataFrame tres veces o usar un bucle).  
Agrega leyenda, título y etiquetas.

```python
especies_unicas = df["species"].unique()

colores_map = {'setosa': 'blue', 'versicolor': 'orange', 'virginica': 'green'}

fig, ax = plt.subplots(figsize=(8, 5))


for nombre_especie, datos in df.groupby('species'):
    ax.scatter(
        datos['sepal_length'], 
        datos['sepal_width'], 
        label=nombre_especie,      # Nombre para la leyenda
        color=colores_map[nombre_especie], # Accedemos al color por el nombre
        alpha=0.7,          
        edgecolors='w'      
    )

ax.set_title("Relación: Longitud vs Ancho del Sépalo", fontsize=14)
ax.set_xlabel("Longitud del Sépalo (cm)")
ax.set_ylabel("Ancho del Sépalo (cm)")
ax.legend() 

plt.show()
```
