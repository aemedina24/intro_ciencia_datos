**P4.** Selecciona solo las columnas `sepal_length` y `species` e imprime las primeras 5 filas.

```python
df_filtrado = df[df["species"] == "virginica"]
print(df_filtrado.head(5))
print("-------------------------------------")
print(f"Numero de filas: {df.shape[0]}")
```
